## 🔧 1. 环境初始化与模块导入

```python
# 环境配置和模块导入
import warnings
warnings.filterwarnings('ignore')

import sys
import os

# 智能路径配置 - 适配VSCode和不同运行环境
current_dir = os.getcwd()
print(f"当前工作目录: {current_dir}")

# 根据当前目录智能添加src路径
if 'notebooks' in current_dir:
    # 从notebooks目录运行
    src_path = os.path.join(os.path.dirname(current_dir), 'src')
else:
    # 从项目根目录运行  
    src_path = os.path.join(current_dir, 'src')

if src_path not in sys.path:
    sys.path.insert(0, src_path)
    
print(f"添加模块路径: {src_path}")

# 导入AI增强房屋市场模型
try:
    from ai_enhanced_housing_model_fixed import (
        SafeAIEnhancedHousingMarketModel,
        run_safe_simulation
    )
    print("✅ AI增强模型导入成功")
except ImportError as e:
    print(f"❌ 模型导入失败: {e}")
    print("请确保ai_enhanced_housing_model_fixed.py文件在src目录中")
    raise

# 导入数据处理和可视化库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 配置中文显示和图表样式
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 8)
sns.set_style("whitegrid")
sns.set_palette("husl")

print("✅ 环境配置完成")
print("✅ 可视化库配置完成")
print("🚀 系统准备就绪，可以开始仿真！")
```

## 🎬 2. 快速演示仿真

```python
# 运行演示仿真
print("🎬 开始AI增强房屋市场仿真演示...")
print("⏳ 请稍候，系统正在初始化智能体和市场环境...")

# 运行仿真
demo_model, demo_results = run_safe_simulation(
    steps=30,           # 仿真步数
    num_residents=50,   # 居民数量
    num_investors=15    # 投资者数量
)

if demo_model and demo_results is not None:
    print("\n🎉 演示仿真成功完成！")
    print(f"📊 收集到 {len(demo_results)} 步仿真数据")
    print(f"🏠 模拟了 {demo_model.num_residents} 个居民和 {demo_model.num_investors} 个投资者")
    
    # 显示数据摘要
    print("\n📈 数据统计摘要:")
    display(demo_results.describe().round(2))
    
    # 显示数据列名
    print("\n📋 收集的指标:")
    for i, col in enumerate(demo_results.columns, 1):
        print(f"{i:2d}. {col}")
else:
    print("❌ 演示仿真失败")
    print("请检查模型文件和依赖包")
```

## 📊 3. 核心数据可视化

```python
# 创建核心指标可视化
if demo_results is not None:
    # 创建2x2子图布局
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('🏠 AI增强房屋市场仿真 - 核心指标分析', fontsize=18, fontweight='bold', y=0.98)
    
    # 1. 房价趋势分析
    ax1 = axes[0,0]
    ax1.plot(demo_results.index, demo_results['Average Property Value'], 
             color='#2E86AB', linewidth=3, marker='o', markersize=5)
    ax1.set_title('💰 平均房价变化趋势', fontsize=14, fontweight='bold', pad=20)
    ax1.set_xlabel('仿真步数', fontsize=12)
    ax1.set_ylabel('房价 (¥)', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.ticklabel_format(style='plain', axis='y')
    
    # 2. 居民满意度变化
    ax2 = axes[0,1]
    ax2.plot(demo_results.index, demo_results['Resident Satisfaction'], 
             color='#A23B72', linewidth=3, marker='s', markersize=5)
    ax2.set_title('😊 居民满意度变化', fontsize=14, fontweight='bold', pad=20)
    ax2.set_xlabel('仿真步数', fontsize=12)
    ax2.set_ylabel('满意度指数', fontsize=12)
    ax2.set_ylim(0, 1)
    ax2.grid(True, alpha=0.3)
    
    # 3. AI市场预测趋势
    ax3 = axes[1,0]
    ax3.plot(demo_results.index, demo_results['AI Market Prediction'], 
             color='#F18F01', linewidth=3, marker='^', markersize=5)
    ax3.axhline(y=0, color='black', linestyle='--', alpha=0.5, linewidth=1)
    ax3.set_title('🤖 AI市场预测信号', fontsize=14, fontweight='bold', pad=20)
    ax3.set_xlabel('仿真步数', fontsize=12)
    ax3.set_ylabel('预测变化率', fontsize=12)
    ax3.grid(True, alpha=0.3)
    
    # 4. AI投资表现
    ax4 = axes[1,1]
    ax4.plot(demo_results.index, demo_results['Investment Performance'], 
             color='#C73E1D', linewidth=3, marker='D', markersize=5)
    ax4.set_title('💼 AI投资策略表现', fontsize=14, fontweight='bold', pad=20)
    ax4.set_xlabel('仿真步数', fontsize=12)
    ax4.set_ylabel('投资表现得分', fontsize=12)
    ax4.grid(True, alpha=0.3)
    
    # 调整布局
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    
    # 保存图表
    os.makedirs('outputs', exist_ok=True)
    plt.savefig('outputs/core_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\n✅ 可视化完成，图表已保存至 outputs/core_metrics.png")
else:
    print("❌ 无法生成可视化 - 缺少仿真数据")
```
