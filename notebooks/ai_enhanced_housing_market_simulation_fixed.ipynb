```python
# ÂàõÂª∫Ê†∏ÂøÉÊåáÊ†áÂèØËßÜÂåñ
if demo_results is not None:
    # ÂàõÂª∫2x2Â≠êÂõæÂ∏ÉÂ±Ä
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('üè† AIÂ¢ûÂº∫ÊàøÂ±ãÂ∏ÇÂú∫‰ªøÁúü - Ê†∏ÂøÉÊåáÊ†áÂàÜÊûê', fontsize=18, fontweight='bold', y=0.98)
    
    # 1. Êàø‰ª∑Ë∂ãÂäøÂàÜÊûê
    ax1 = axes[0,0]
    ax1.plot(demo_results.index, demo_results['Average Property Value'], 
             color='#2E86AB', linewidth=3, marker='o', markersize=5)
    ax1.set_title('üí∞ Âπ≥ÂùáÊàø‰ª∑ÂèòÂåñË∂ãÂäø', fontsize=14, fontweight='bold', pad=20)
    ax1.set_xlabel('‰ªøÁúüÊ≠•Êï∞', fontsize=12)
    ax1.set_ylabel('Êàø‰ª∑ (¬•)', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.ticklabel_format(style='plain', axis='y')
    
    # 2. Â±ÖÊ∞ëÊª°ÊÑèÂ∫¶ÂèòÂåñ
    ax2 = axes[0,1]
    ax2.plot(demo_results.index, demo_results['Resident Satisfaction'], 
             color='#A23B72', linewidth=3, marker='s', markersize=5)
    ax2.set_title('üòä Â±ÖÊ∞ëÊª°ÊÑèÂ∫¶ÂèòÂåñ', fontsize=14, fontweight='bold', pad=20)
    ax2.set_xlabel('‰ªøÁúüÊ≠•Êï∞', fontsize=12)
    ax2.set_ylabel('Êª°ÊÑèÂ∫¶ÊåáÊï∞', fontsize=12)
    ax2.set_ylim(0, 1)
    ax2.grid(True, alpha=0.3)
    
    # 3. AIÂ∏ÇÂú∫È¢ÑÊµãË∂ãÂäø
    ax3 = axes[1,0]
    ax3.plot(demo_results.index, demo_results['AI Market Prediction'], 
             color='#F18F01', linewidth=3, marker='^', markersize=5)
    ax3.axhline(y=0, color='black', linestyle='--', alpha=0.5, linewidth=1)
    ax3.set_title('ü§ñ AIÂ∏ÇÂú∫È¢ÑÊµã‰ø°Âè∑', fontsize=14, fontweight='bold', pad=20)
    ax3.set_xlabel('‰ªøÁúüÊ≠•Êï∞', fontsize=12)
    ax3.set_ylabel('È¢ÑÊµãÂèòÂåñÁéá', fontsize=12)
    ax3.grid(True, alpha=0.3)
    
    # 4. AIÊäïËµÑË°®Áé∞
    ax4 = axes[1,1]
    ax4.plot(demo_results.index, demo_results['Investment Performance'], 
             color='#C73E1D', linewidth=3, marker='D', markersize=5)
    ax4.set_title('üíº AIÊäïËµÑÁ≠ñÁï•Ë°®Áé∞', fontsize=14, fontweight='bold', pad=20)
    ax4.set_xlabel('‰ªøÁúüÊ≠•Êï∞', fontsize=12)
    ax4.set_ylabel('ÊäïËµÑË°®Áé∞ÂæóÂàÜ', fontsize=12)
    ax4.grid(True, alpha=0.3)
    
    # Ë∞ÉÊï¥Â∏ÉÂ±Ä
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    
    # ‰øùÂ≠òÂõæË°®
    os.makedirs('outputs', exist_ok=True)
    plt.savefig('outputs/core_metrics.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\n‚úÖ ÂèØËßÜÂåñÂÆåÊàêÔºåÂõæË°®Â∑≤‰øùÂ≠òËá≥ outputs/core_metrics.png")
else:
    print("‚ùå Êó†Ê≥ïÁîüÊàêÂèØËßÜÂåñ - Áº∫Â∞ë‰ªøÁúüÊï∞ÊçÆ")
```
